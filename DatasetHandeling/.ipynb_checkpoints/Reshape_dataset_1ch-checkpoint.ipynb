Expirement with different dataset variations. This notebook loads the datasets from the Crate_dataset notebook and reshapes it. 
This notebook is obsolete for making training sets. Use it for creating test sets that are to be used many times as it is more efficient than completing the creation process for each experiment

```python
import numpy as np
import matplotlib.pyplot as plt
filn = 7
loc = 'Skogen'
datasetDir = '../../Datasets/DVRUP_1f_UINT8_82020/'
datasetName = loc + str(filn) +'DUINT8'
outDatasetDir = '../../Datasets/DVRUP_1f_2f2b/'
outDatasetName = loc + str(filn) +'_2a2b_'
keepEach = 5
framesBefore = 2
framesAfter = 2
chan = 1
frames = framesBefore + 1 + framesAfter
```

```python
X = np.load(datasetDir + datasetName + 'X.npy')
Y = np.load(datasetDir + datasetName + 'Y.npy')
```

```python
print(X.shape)
print(Y.shape)
X_r = X[:, :, :, 0]
X_g = X[:, :, :, 1]
X_b = X[:, :, :, 2]

Y_r = Y[:, :, :, 0]
Y_g = Y[:, :, :, 1]
Y_b = Y[:, :, :, 2]

print(X_r.shape)
```

```python
def getFramesAround(DS, num, fbf = 1, faf = 1):
    """ Returns a matrix with the fbf frames before and the faf frames after frame number num """
    return np.stack(DS[(num-fbf):(1+num+faf)], axis = 2)

def reshapeDStoFBF(X, Y, fbf = 1, faf = 1):
    """ Reshapes the dataset so that each entry contains the fbf frames before and faf frames after the entry """
    #Y = Y[framesBefore:(-1 * framesAfter)]
    rn = framesBefore
    buf =  np.empty((len(X) - (framesAfter + framesBefore), len(X[0]), len(X[0][0]),frames), np.dtype('uint8'))
    bufy = np.empty((len(Y) - (framesAfter + framesBefore), len(Y[0]), len(Y[0][0]),frames), np.dtype('uint8'))
    for i in range(len(X) - (framesAfter + framesBefore)):
        buf[rn - framesBefore] = getFramesAround(X, rn,fbf,faf)
        bufy[rn - framesBefore] = getFramesAround(Y, rn,fbf,faf)
        rn += 1
    print('Reshaped dataset to size: ', buf.shape, ' Y:', bufy.shape)
    return buf, bufy
```

```python
X_r, Y_r = reshapeDStoFBF(X_r, Y_r, framesBefore, framesAfter)
X_g, Y_g = reshapeDStoFBF(X_g, Y_g, framesBefore, framesAfter)
X_b, Y_b = reshapeDStoFBF(X_b, Y_b, framesBefore, framesAfter)
```

```python
def showFrame(frame, title = 'Frame', show = True):
    plt.imshow(frame)
    plt.title(title)
    if show:
        plt.show()
```

```python
frame = 100
c = 2
showFrame(X_r[frame, :, :, c])
showFrame(Y_r[frame, :, :, c])
showFrame(X_g[frame, :, :, c])
showFrame(Y_g[frame, :, :, c])
showFrame(X_b[frame, :, :, c])
showFrame(Y_b[frame, :, :, c])
```

```python
def keepEach(X, Y, num):
    """ Returns new x and y matreces that only contains evry num frame of the originals """
    t = 1
    a = 0
    nX = np.empty((int(len(X)/num), len(X[0]), len(X[0][0]), len(X[0][0][0])), np.dtype('uint8'))
    nY = np.empty((int(len(Y)/num), len(Y[0]), len(Y[0][0]), len(Y[0][0][0])), np.dtype('uint8'))
    for i in range(len(X)):
        if (t == num):
            nX[a] = X[i]
            nY[a] = Y[i]
            t = 1
            a += 1
        else:
            t += 1
    return nX, nY

print(X_r.shape, Y_r.shape)
#X_r, Y_r = keepEach(X_r, Y_r, 5)
#X_g, Y_g = keepEach(X_g, Y_g, 5)
#X_b, Y_b = keepEach(X_b, Y_b, 5)
```
