Expirement with different dataset variations. This notebook loads the datasets from the Crate_dataset notebook and reshapes it. 


```python
import numpy as np
import matplotlib.pyplot as plt
datasetDir = '../../DATASET/'
datasetName = 'Hagen1DUINT8'
outDatasetDir = '../DATASET5thF2b2a/'
outDatasetName = 'Hagen5th2a2b'
keepEach = 5
framesBefore = 2
framesAfter = 2
chan = 3
frames = framesBefore + 1 + framesAfter
```

```python
X = np.load(datasetDir + datasetName + 'X.npy')
Y = np.load(datasetDir + datasetName + 'Y.npy')
```

```python
print(X.shape)
print(Y.shape)
```

```python
def getFramesAround(DS, num, fbf = 1, faf = 1):
    """ Returns a matrix with the fbf frames before and the faf frames after frame number num """
    nDS = DS[num, :, :, :]
    for i in range(fbf):
        nDS = np.append( DS[num - (i + 1), :, :, :], nDS, axis = 2)
    for i in range(faf):
        nDS = np.append( nDS, DS[num + (i + 1), :, :, :], axis = 2)
    return nDS

def reshapeDStoFBF(X, Y, fbf = 1, faf = 1):
    """ Reshapes the dataset so that each entry contains the fbf frames before and faf frames after the entry """
    Y = Y[framesBefore:(-1 * framesAfter)]
    rn = framesBefore
    buf = np.empty((len(X) - (framesAfter + framesBefore), len(X[0]), len(X[0][0]), chan * frames), np.dtype('uint8'))
    for i in range(len(X) - (framesAfter + framesBefore)):
        buf[rn - framesBefore] = getFramesAround(X, rn,fbf,faf)
        rn += 1
    print('Reshaped dataset to size: ', buf.shape, ' Y:', Y.shape)
    return buf, Y
```

```python
X, Y = reshapeDStoFBF(X, Y, framesBefore, framesAfter)
```

```python
def showFrame(frame, title = 'Frame', show = True):
    plt.imshow(frame)
    plt.title(title)
    if show:
        plt.show()
```

```python
frame = 400
for i in range(frames):
    showFrame(X[frame, :, :, (i*3):(i*3) + 3])
showFrame(Y[frame])
```

```python
def keepEach(X, Y, num):
    """ Returns new x and y matreces that only contains evry num frame of the originals """
    t = 1
    a = 0
    nX = np.empty((int(len(X)/num), len(X[0]), len(X[0][0]), len(X[0][0][0])), np.dtype('uint8'))
    nY = np.empty((int(len(Y)/num), len(Y[0]), len(Y[0][0]), len(Y[0][0][0])), np.dtype('uint8'))
    for i in range(len(X)):
        if (t == num):
            nX[a] = X[i]
            nY[a] = Y[i]
            t = 1
            a += 1
        else:
            t += 1
    return nX, nY

nX, nY = keepEach(X, Y, 5)
```

```python
print(nX.shape, nY.shape)
```

```python
frame = 100
for i in range(frames):
    showFrame(nX[frame, :, :, (i*3):(i*3) + 3])
showFrame(nY[frame])
```

```python
np.save(outDatasetDir + outDatasetName + 'X', nX)
np.save(outDatasetDir + outDatasetName + 'Y', nY)
```
